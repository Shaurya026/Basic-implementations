## Linear Regression : 

```python
import numpy as np 

def function(num_iterations,lr):
    fun = lambda x : x**2 + 2*x + 12
    x = np.random.randint(0,20,1000)
    y = fun(x)
    
    number_of_iter = num_iterations #1000

    w = 0
    b = 0
    for i in range(number_of_iter) :
        z = w*x + b 
        cost = (1/len(x)) * np.sum( (y-z)**2 )
        dz = -2*(y-z)
        dw = (1/len(x)) * np.sum(x*dz)                                 
        db = (1/len(x)) * np.sum(dz)
        w = w - lr*dw
        b = b - lr*db
        if i % 100 == 0 :
            print('cost after',i,'th iteration:',cost)
    x =  int(input('enter the number on which function prediction needs to be made :'))
    print('output :',(w*x + b))
    print('actual value :',fun(x))
```

```python
function(1000,0.01)
```

```python
function(1000,0.001)
```

```python
function(1000,0.0001)
```

```python
function(1000,0.003)
```

## Normalizing to make the function fall to the minima faster : 

```python
def function(num_iterations,lr):
    fun = lambda x : x**2 + 2*x + 12
    x = np.random.randint(0,20,1000)
    y = fun(x)
    
    val_x = [np.mean(x),np.std(x)]
    val_y = [np.mean(y),np.std(y)]
    
    #print(val_x,val_y)
    
    x = (x)/val_x[1]
    #y = (y )/np.std(y)
    number_of_iter = num_iterations #1000

    w = 0
    b = 0
    for i in range(number_of_iter+1) :
        z = w*x + b 
        cost = (1/(2*len(x))) * np.sum( (y-z)**2 )
        dz = -1*(y-z)
        dw = (1/len(x)) * np.sum(x*dz)                                 
        db = (1/len(x)) * np.sum(dz)
        w = w - lr*dw
        b = b - lr*db
        if i % 100 == 0 :
            print('cost after',i,'th iteration:',cost)
    x =  int(input('enter the number on which function prediction needs to be made :'))
    print('output :',(w*(x/val_x[1]) + b) )
    print('actual value :',fun(x))
```

```python
function(10000,0.1)
```

## Trying with Quadratic feature : 
i.e. with two input features : 

```python
import numpy as np 
```

```python
def function(num_iterations,lr):
    fun = lambda x : x**2 + 2*x + 12
    x1 = np.random.randint(0,20,1000)
    x2 = x1 ** 2
    y = fun(x1)
    
    val_x1 = [np.mean(x1),np.std(x1)]
    val_x2 = [np.mean(x2),np.std(x2)]
    val_y = [np.mean(y),np.std(y)]
    
    #print(val_x,val_y)
    
    x1 = (x1-val_x1[0])/val_x1[1]
    x2 = (x2-val_x2[0])/val_x2[1]
    #y = (y )/np.std(y)
    number_of_iter = num_iterations #1000
    
    # input feature vector : 
    x = np.array([x1,x2])
    w = np.zeros((2,1))
    b = 0
    for i in range(number_of_iter+1) :
        z = np.dot(w.T,x) + b 
        cost = (1/(2*z.shape[1])) * np.sum( (z-y)**2 )
        dz =  (z-y)
        dw1 = (1/z.shape[1]) * np.sum(x1*dz)                                 
        dw2 = (1/z.shape[1])* np.sum(x2*dz)                                 
        db = (1/z.shape[1]) * np.sum(dz)
        w[0] = w[0] - lr*dw1
        w[1] = w[1] - lr*dw2
        b = b - lr*db
        if i % 100 == 0 :
            print('cost after',i,'th iteration:',cost)
    x =  int(input('enter the number on which function prediction needs to be made :'))
    x_mult = np.array([(x-val_x1[0])/val_x1[1],((x**2)-val_x2[0])/val_x2[1]])
    print('output :',(np.dot(w.T,x_mult) + b))
    print('actual value :',fun(x))
    print(w.T)
```

```python
function(1000,0.1)
```

## Making the same changes in Keras by giving it 2 inputs : 

```python
import tensorflow as tf 
from tensorflow import keras
```

```python
model = tf.keras.Sequential([keras.layers.Dense(units = 1, input_shape = (2,))])
opt = tf.keras.optimizers.SGD(learning_rate=0.00001)
model.compile(optimizer=opt,loss='mean_squared_error')
fun = lambda x : x**2 + 2*x + 12
x1 = np.random.randint(0,20,1000)
x2 = x1 ** 2
x = np.array([x1,x2])
y = fun(x1)
x = x.T
y = y.reshape((1000,1))
#print(y)
model.fit(x,y,epochs=1000)
```
